# Now build the model to predict the movie ratings

```python
########################################################################################################
# prepare for building recommendation system - store the number of users and movies in variables
########################################################################################################
n_users = Adata.user_id.nunique()
n_movies = len(Movie_rate_mean.index)
```

```python
########################################################################################################
# print the number of users and movies
########################################################################################################
print(n_users)
print(n_movies)
```

```python
########################################################################################################
# define a function to reformat the userid data to numeric data
########################################################################################################
userid_list = Adata.user_id.unique()
print(" Length of user_id List:", len(userid_list))
def get_user_id_numeric_id(user_id):

    itemindex = np.asarray(userid_list==user_id).nonzero()
    return itemindex[0][0]
```

```python
movie_list = Movie_rate_mean.index
print(" Length of movie List:", len(movie_list))
def get_movie_numeric_id(movie):

    itemindex = np.asarray(movie_list==movie).nonzero()
    return itemindex[0][0]
```

```python
########################################################################################################
# create a new column with the reformated user_id
########################################################################################################
Adata['user_id_order'] = Adata['user_id'].apply(get_user_id_numeric_id)
```

```python
########################################################################################################
# remove the old char. userid column and realign the columns to bring the newly created user_id first
########################################################################################################
Adata1 = Adata['user_id_order']
Adata2 = Adata.iloc[0:,1:207]
```

```python
########################################################################################################
# concatenate the above created fields so that user_id_order column is in the begining
########################################################################################################
Adatafull = pd.concat([Adata1,Adata2],axis=1)
```

```python
Adatafull.head()
```

```python
Adatamelt = Adatafull.melt(id_vars='user_id_order',var_name='movie',value_name='ratings')
```

```python
########################################################################################################
# create a new column with the reformated user_id
########################################################################################################
Adatamelt['movie_id_order'] = Adatamelt['movie'].apply(get_movie_numeric_id)
```

```python
########################################################################################################
# create a new dataframe with new userid, new movie id and the ratings 
########################################################################################################
Adatamelt1 = Adatamelt['user_id_order']
Adatamelt2 = Adatamelt['movie_id_order']
Adatamelt3 = Adatamelt['ratings']
Adatacomb = pd.concat([Adatamelt1,Adatamelt2,Adatamelt3],axis=1)
Adatacomb
```

```python
########################################################################################################
# Load the dataframe back into the original dataframe
########################################################################################################
Adata = Adatacomb
```

# Now split the file into train and test files

```python
########################################################################################################
# split the data into train and test files
########################################################################################################
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(Adata, test_size=0.20)
```

```python
########################################################################################################
# check the bottom five rows of test dataset
########################################################################################################
test_data.tail()
```

```python
########################################################################################################
# Create two user-movie matrices, one for training and another for testing- below is for training
########################################################################################################

train_data_matrix = np.zeros((n_users, n_movies))
   
for line in train_data.itertuples():
 
     train_data_matrix[line[1], line[2]] = line[3] 

```

```python
########################################################################################################
# Create the user-movie matrix for testing
########################################################################################################

test_data_matrix = np.zeros((n_users, n_movies))

for line in test_data.itertuples():

    test_data_matrix[line[1], line[2]] = line[3]
```

```python
########################################################################################################
# calculate the user similarity pairwise distances using cosine metrics using the train_datset
########################################################################################################
from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
```

```python
########################################################################################################
# check the user similarity array  
########################################################################################################
user_similarity
```

```python
########################################################################################################
# chek the saphe of the user similarity matrixs
########################################################################################################
user_similarity.shape
```

```python
########################################################################################################
# create a function to predict the ratings
########################################################################################################
def predict(ratings, similarity):
   
        mean_user_rating = ratings.mean(axis=1)
        
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis]) 
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
   
        return pred
```

```python
########################################################################################################
# predict on the test dataset
########################################################################################################
user_prediction = predict(test_data_matrix, user_similarity)
```

```python
########################################################################################################
# physically look at  the first 10 predictions
########################################################################################################
user_prediction[test_data_matrix.nonzero()][0:10]
```

```python
########################################################################################################
# physically look at  the first 10 true values 
########################################################################################################
test_data_matrix[test_data_matrix.nonzero()][0:10]
```

```python
########################################################################################################
# Load the required packages and define a function to calculate the root menn squared error
########################################################################################################
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten() 
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))
```

```python
########################################################################################################
# prin the root mean squared error from the predictions
########################################################################################################
print('User-based CF RMSE: ' + str(rmse(user_prediction, test_data_matrix)))
```
